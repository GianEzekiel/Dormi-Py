```python
# Define user and admin account files
user_file = "users.txt"
admin_file = "admins.txt"

# Define user and admin account dictionaries
users = {}
admins = {}

# Load user and admin accounts from files
def load_accounts():
    with open(user_file, "r") as f:
        for line in f:
            username, password = line.strip().split(":")
            users[username] = password
    with open(admin_file, "r") as f:
        for line in f:
            username, password = line.strip().split(":")
            admins[username] = password

# Save user and admin accounts to files
def save_accounts():
    with open(user_file, "w") as f:
        for username, password in users.items():
            f.write(f"{username}:{password}\n")
    with open(admin_file, "w") as f:
        for username, password in admins.items():
            f.write(f"{username}:{password}\n")

# Create a new user account
def create_user(username, password):
    if username in users or username in admins:
        print("Username already exists")
    else:
        users[username] = password
        save_accounts()
        print("User account created successfully")

# Create a new admin account
def create_admin(username, password):
    if username in users or username in admins:
        print("Username already exists")
    else:
        admins[username] = password
        save_accounts()
        print("Admin account created successfully")

# Authenticate a user or admin account
def authenticate(username, password):
    if username in users and users[username] == password:
        print("User account authenticated")
    elif username in admins and admins[username] == password:
        print("Admin account authenticated")
    else:
        print("Invalid username or password")

# Example usage
load_accounts()
create_user("john", "password123")
create_admin("jane", "adminpass456")
authenticate("john", "password123")
authenticate("jane", "adminpass456")
```
